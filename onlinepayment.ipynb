```python
import pandas as pd
import numpy as np
data = pd.read_csv("onlinefraud.csv")
data.head()
```

```python
data.type.value_counts()
```

```python
type = data["type"].value_counts()
transactions = type.index
quantity = type.values

import plotly.express as px
figure = px.pie(data,
             values=quantity,
             names=transactions,hole = 0.5,
             title="Distribution of Transaction Type")
figure.show()
```

```python
correlation = data.select_dtypes(include=np.number).corr() # Select only numerical columns for correlation calculation
print(correlation["isFraud"].sort_values(ascending=False))
```

```python
data["type"] = data["type"].map({"CASH_OUT": 1, "PAYMENT": 2,
                                 "CASH_IN": 3, "TRANSFER": 4,
                                 "DEBIT": 5})
data["isFraud"] = data["isFraud"].map({0: "No Fraud", 1: "Fraud"})
data.head()
```

```python
from sklearn.model_selection import train_test_split
x = np.array(data[["type", "amount", "oldbalanceOrg", "newbalanceOrig"]])
y = np.array(data[["isFraud"]])
```

```python
# training a machine learning model
from sklearn.tree import DecisionTreeClassifier
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.20, random_state=42)
model = DecisionTreeClassifier()
# Convert ytrain to numerical values (0 or 1)
ytrain_numeric = np.where(ytrain == 'Fraud', 1, 0)
model.fit(xtrain, ytrain_numeric)
# Convert ytest to numerical values (0 or 1)
ytest_numeric = np.where(ytest == 'Fraud', 1, 0)
model.score(xtest, ytest_numeric)
```

```python
features = np.array([[4, 9000.60, 9000.60, 0.0]])
print(model.predict(features))
```
